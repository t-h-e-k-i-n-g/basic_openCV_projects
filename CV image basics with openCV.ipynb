```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
black_img = np.zeros([512,512,3], dtype = np.int16)
plt.imshow(black_img)
```

```python
# drawing a circle
cv2.circle(black_img, (400,100), 50, color= (255,0,0), thickness = 3)
plt.imshow(black_img)
```

```python
# drawing a filled circle
cv2.circle(black_img, (400,200), 50, color= (0,255,0), thickness = -1)
plt.imshow(black_img)
```

```python
# drawing a rectangle
cv2.rectangle(black_img, pt1 = (100,100),
             pt2= (200,200), color = (0,0,255), thickness = 3)
plt.imshow(black_img)
```

```python
# drawing a filled rectangle
cv2.rectangle(black_img, pt1 = (300,300),
             pt2= (350,350), color = (0,255,255), thickness = -1)
plt.imshow(black_img)
```

```python
# drawing a triangle
pts = np.array([[100,300],[200,300],[250,400]], np.int32)
#pts = vertices.reshape(-1,1,2)
cv2.polylines(black_img, [pts],
              isClosed = True,
             color= (255,255,0),
             thickness = 3)
plt.imshow(black_img)
```

```python
# drawing a filled triangle
vert = np.array([[150,350],[200,350],[150,500]], np.int32)
cv2.fillPoly(black_img, [vert],
            color = (120,120,120))
plt.imshow(black_img)
```

```python
# writing text on image
font = cv2.FONT_HERSHEY_COMPLEX
text = 'bts'
cv2.putText(black_img, text, org=(400,450),
           fontFace = font, fontScale = 2,
           color = (35,67,120), thickness = 3,
           lineType = cv2.LINE_4)
plt.imshow(black_img)
```

```python
# draw a line
cv2.line(black_img, pt1=(250,0), 
       pt2=(250,500),
       color=(30,88,90),
       thickness = 5)
plt.imshow(black_img)
```
